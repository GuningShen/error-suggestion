```python
import json
import os
import pandas as pd
from datetime import datetime 
```

## First step: Read the raw JSON file

```python
data_folder = "../logs/"
output_dir = "output_data/"


# Initialize a list to hold concatenated data
concatenated_data = []

# Iterate through each file in the data folder
for file in os.listdir(data_folder):
    if file.endswith(".json"):
        # Construct full file path
        file_path = os.path.join(data_folder, file)
        # Open and read the JSON file
        with open(file_path, "r") as f:
            data = json.load(f)
            # Append the data to the concatenated data list
            concatenated_data.append(data)

concatenated_data[0:1]

# data_folder = "../logs/"
# output_dir = "output_data/"
# file = "zwy324.json"
# raw_json = None

# #Handy for making an output dir inline if needed. 
# if not os.path.isdir(output_dir):
#     os.makedirs(output_dir)

# #I always read JSON in as JSON first and then parse. 
# with open(data_folder + file, "r+") as f:
#     raw_json = json.load(f)
# raw_json
```

```python
raw_records = []
for raw_json in concatenated_data:   
    user_name = raw_json['session_id']
    log_id = raw_json['log_id']
    log_type = raw_json['log_type']
    for i in raw_json['log']['commands']:
            to_add = {
                "Log ID": log_id, 
                "Username": user_name, 
                "Log Type": log_type,
            }
            
            #Add each command to array and update 
            to_add.update(i)
            
            raw_records.append(to_add)

#Look at the records we have made
raw_records[0:1]
```

```python
# Convert to a DataFrame
pd_records = pd.DataFrame.from_records(raw_records)
pd_records.head()
```

```python
# Clean up
pd_records.drop(columns=['diff_l', 'diff_wc', 'new_l', 'new_wc', 'old_l', 'old_wc', "total_time(s)"], inplace=True) 
pd_records.head()
```
